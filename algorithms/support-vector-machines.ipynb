Les machines à vecteurs de support (SVM) sont des algorithmes d'apprentissage supervisé utilisés principalement pour la classification et la régression. L'objectif principal des SVM est de trouver un hyperplan (une frontière de décision) qui sépare au mieux les données en deux catégories, maximisant ainsi la marge entre les points de données les plus proches de chaque classe.

Les SVM sont particulièrement efficaces pour traiter des ensembles de données complexes et non linéaires grâce à des techniques de transformation de l'espace, telles que le noyau. Les noyaux permettent de projeter les données dans un espace de dimension supérieure où une séparation linéaire est possible.

Les SVM sont robustes face aux problèmes de surapprentissage grâce à leur capacité à maximiser la marge entre les classes, ce qui limite la généralisation excessive. Ils sont également efficaces avec des ensembles de données de petite à moyenne taille.

En résumé, les SVM sont des outils puissants pour la classification et la régression, adaptés à une variété de domaines tels que la reconnaissance de formes, l'analyse d'image, la classification de texte, et bien d'autres, en trouvant un hyperplan optimal pour séparer les données en classes distinctes tout en maximisant la marge.

## Données

Création d'un jeu de données synthétiques:

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

random_state=123

# Créer les données
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    n_clusters_per_class=2,
    class_sep=.9,
    random_state=random_state,
)

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

# Vérifier les dimensions (division 80%-20%)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Visualisation:

```python
import matplotlib.pyplot as plt 
%matplotlib inline

plt.scatter(X[:,0], X[:,1], alpha=0.5, c=y, cmap=plt.cm.rainbow)
```

## Machines à Vecteurs de Support avec SciKit-Learn

Docs: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

clf = SVC(random_state=random_state)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("Train accuracy:", accuracy_score(y_train, clf.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, clf.predict(X_test)))
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, y, clf=clf, legend=2) # X_highlight=X_test)
plt.xlabel('X0')
plt.ylabel('X1')
plt.title('Machines à Vecteurs de Support')
plt.show()
```
